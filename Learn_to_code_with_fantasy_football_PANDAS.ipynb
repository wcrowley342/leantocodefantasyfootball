**Summary Functions on Boolean Columns**

```python
pg = pd.read_csv('https://raw.githubusercontent.com/wcrowley342/leantocodefantasyfootball/main/player_game_2017_sample.csv')
```

```python
pg['good_rb_game'] = ((pg['pos'] == 'RB')& 
                      (pg['rush_yards'] >=100))

pg['good_rb_game'].mean()
```

```python
pg['good_rb_game'].sum()
```

```python
#did anyone pass for over 400 yards?

(pg['pass_yards'] > 400).any()
```

```python
(pg[['rush_yards', 'rec_yards']] > 50).any(axis=1)
```

```python
#call a function on this column to see how often it happened
(pg[['rush_yards', 'rec_yards']] > 50).any(axis=1).sum()
```

```python
#what about 50 rushing AND recieving yards? use all, instead of any

(pg[['rush_yards', 'rec_yards']] > 50).all(axis=1).sum()
```

**Value counts functions**

```python
#value counts, summarizes the frequency of individual values
pg['pos'].value_counts()
```

```python
#normalizing / getting percentages of them so they add up to 1 and represent proportions

pg['pos'].value_counts(normalize=True)
```
